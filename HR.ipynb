```python
df.drop(['EmployeeNumber','DailyRate', 'HourlyRate', 'MonthlyRate'], axis = 1, inplace = True)
```

```python
df_cat = df[['Attrition', 'BusinessTravel','Department',
                       'EducationField','Gender','JobRole',
                       'MaritalStatus',
                       'Over18', 'OverTime']].copy()
```

```python
df_cat.head()
```

```python
df_num = df[['Age','DistanceFromHome','EnvironmentSatisfaction','JobInvolvement','JobLevel','JobSatisfaction', 'RelationshipSatisfaction', 'StockOptionLevel','TrainingTimesLastYear']].copy()
```

```python
df_num.head()
```

```python
Num_val = {'Yes':1, 'No':0}
df_cat['Attrition'] = df_cat["Attrition"].apply(lambda x: Num_val[x])
df_cat.head()
```

```python
df_cat = pd.get_dummies(df_cat)
df_cat.head()
```

```python
dff = pd.concat([df_num, df_cat], axis=1)
dff.head()
```

```python
x = dff.drop('Attrition',axis=1)
```

```python
y = dff.Attrition
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
#create the train/test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=10)
#Create the model and train
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
train_pred = rfc.predict(x_train)
#predict the results for test
test_pred = rfc.predict(x_test)
#test the accuracy
a1=accuracy_score(y_train, train_pred)
a2=accuracy_score(y_test, test_pred)

print("train score=", a1)
print("test score=", a2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict for test dataset
y_pred = clf.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.externals import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(rfc, 'hrmodel.pkl') 
  
# Load the model from the file 
rfc_from_joblib = joblib.load('hrmodel.pkl')  
  
# Use the loaded model to make predictions 
rfc_from_joblib.predict(x_test) 
```
